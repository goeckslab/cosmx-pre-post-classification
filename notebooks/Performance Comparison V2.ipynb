```python
import pandas as pd
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
radii = [0,15,30,60, 90, 120]
```

```python
data_frames = []
for radius in radii:
    load_path = Path("v2", str(radius))
    patient_2: pd.DataFrame = pd.read_csv(Path(load_path, "9_2_metrics.csv"))
    patient_2["Patient"] = "Patient 2"
    patient_2["Radius"] = radius
    data_frames.append(patient_2)
    
    patient_3 = pd.read_csv(Path(load_path, "9_3_metrics.csv"))
    patient_3["Patient"] = "Patient 3"
    patient_3["Radius"] = radius
    data_frames.append(patient_3)

    patient_14 = pd.read_csv(Path(load_path, "9_14_metrics.csv"))
    patient_14["Patient"] = "Patient 14"
    patient_14["Radius"] = radius
    data_frames.append(patient_14)

    patient_15 = pd.read_csv(Path(load_path, "9_15_metrics.csv"))
    patient_15["Patient"] = "Patient 15"
    patient_15["Radius"] = radius
    data_frames.append(patient_15)


metrics = pd.concat(data_frames, axis=0)
metrics = metrics.drop(columns=["Unnamed: 0"])
```

```python
metrics
```

```python
sns.lineplot(data=metrics, x="Radius", y="Accuracy", hue="Patient")
plt.savefig("accuracy_per_radius_v2.png", dpi=300)
```
